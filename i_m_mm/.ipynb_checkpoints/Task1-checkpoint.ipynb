```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
sigma = [[1, 3], [3, 5]]
#
#sigma = [[1,0], [0, 2]]
```

```python
eigen_vectors = np.linalg.eig(sigma)
```

```python
eigen_vectors[1]
```

```python
x = np.linspace(-1, 1, 100)
y =  np.sqrt(1-np.square(x)) 

    
    
coords = np.zeros((200, 2))

for i in range(100):
    coords[i] = np.dot( sigma, [x[i], y[i]])
    coords[i+100] = np.dot(  sigma, [x[i], -y[i]])
```

```python
origin = [0], [0 ]

plt.quiver(*origin,eigen_vectors[1][:,0],eigen_vectors[1][:,1], scale = 4)
#plt.quiver( eigen_vectors[1][:,0],eigen_vectors[1][:,1], scale = 4)

#plt.plot(x, y)
plt.plot(coords[:, 0], coords[:,1])
```
