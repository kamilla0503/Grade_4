```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.path import Path
from scipy.spatial import Delaunay
import scipy.optimize
import scipy.spatial

import pandas as pd
```

```python
def line(p1,p2,x):
    return p1[1]+(p2[1]-p1[1])*((x-p1[0])/(p2[0]-p1[0]))

def create_test_points(points_bounds, segs=30, delta=5):
    X,Y = np.meshgrid(
        np.linspace(points_bounds[1,0] + delta, points_bounds[0,0] - delta, segs), 
        np.linspace(points_bounds[1,1] + delta, points_bounds[0,1] - delta, segs)
    )
    inner = np.vstack((X.ravel(),Y.ravel())).T
    return inner


def in_border(p, pb, delta=1):
    mask = p[:,1] < line(pb[0], pb[1], p[:,0]) + delta 
    return mask

points_bounds = np.array([
    [10  , 120/4],
    [110/3, 10],
    [10,  10],
    [10, 120/4]
])

test_points = create_test_points(points_bounds, segs=15, delta=0)
test_points = test_points[in_border(test_points, points_bounds, delta=-0.1)]

w, h = plt.figaspect(1.3) * 2
plt.figure(figsize=(w, h))
plt.plot(points_bounds[:,0],points_bounds[:,1],'.-r')
plt.plot(test_points[:,0],test_points[:,1],'.b')
plt.xlabel('R')
plt.ylabel('L')
plt.title('Облако точек допустимых значений')
```

```python
def detail_main_parts(R,L):
    g = 3.0
 
 
    h = 0.1
    
    W=150 
    
    G0y =  np.arange(0, W,  g)
    G0x = np.zeros_like(G0y)

    G1x =  np.arange(0, W+h,  g)
    G1y = np.zeros_like(G1x)+W
    
    G2y =  np.arange(0, W,  g)
    G2x = np.zeros_like(G2y)+W
    
    
    a=(W-3*L)/2.0
    b = 150-25-R
    
    D0x = np.arange(0, L,  g)
    D0y = np.zeros_like(D0x) 
    
    D1x = np.arange(0, L+h,  g)+L+a
    D1y = np.zeros_like(D1x)     

    D2x = np.arange(0, L,  g)+L+a+L+a
    D2y = np.zeros_like(D2x) 
    
    U0y =  np.arange(0+h, b,  g)
    U0x = np.zeros_like(U0y)+L  
    
    U1x = np.zeros_like(U0y)+L+a
    U2x = np.zeros_like(U0y)+L +a +L 

    U3x = np.zeros_like(U0y)+L +a +L + a 
    
    U4x = np.zeros_like(U0y)+L +a +L +a+L
 
    M0x = np.arange(0, a-2*R,  g) + L + R 
    M0y = np.zeros_like(M0x)+W-25 
    
    M1x = np.arange(0, a-2*R,  g) +W - (a-2*R)-L-R
    M1y = np.zeros_like(M1x)+W-25
    
    n_p = np.pi*R/2/g
    
    a_c = np.linspace(0, np.pi/2.0,  n_p+1)
    
    
    C0x = R*np.cos(a_c+np.pi/2.0)+L+R
    C0y = R*np.sin(a_c+np.pi/2.0)+W-25   -R

    C1x = R*np.cos(a_c )+L+a -R
    C1y = R*np.sin(a_c )+W-25 -R  
    
    C2x = R*np.cos(a_c+np.pi/2.0)+L+R+a+L
    C2y = R*np.sin(a_c+np.pi/2.0)+W-25   -R

    C3x = R*np.cos(a_c )+ W - L-R
    C3y = R*np.sin(a_c )+W-25 -R      
    
    
    fixed_border = np.vstack((np.vstack((G0x, G0y)).T,np.vstack((G1x, G1y)).T, np.vstack((G2x, G2y)).T )) 
 
    down = np.vstack((np.vstack((D0x, D0y)).T,np.vstack((D1x, D1y)).T, np.vstack((D2x, D2y)).T )) 
    
    up =  np.vstack((np.vstack((U0x, U0y)).T,np.vstack((U1x, U0y)).T,np.vstack((U2x, U0y)).T,np.vstack((U3x, U0y)).T,np.vstack((U4x, U0y)).T  ))
    
    middle = np.vstack((np.vstack((M0x, M0y)).T, np.vstack((M1x, M1y)).T))
    
    circle = np.vstack((np.vstack((C0x, C0y)).T,np.vstack((C1x, C1y)).T,np.vstack((C2x, C2y)).T,np.vstack((C3x, C3y)).T  )) 
    
    GNodes =  np.vstack((fixed_border, down, up, middle, circle ))
    
   
    margin = 2.4
    
    x = np.arange(margin, 150 ,  margin)
    y = np.arange(margin, 150 ,  margin)
    X, Y = np.meshgrid(x, y)
    INodes = np.vstack((X.ravel(), Y.ravel())).T
    mar = 0.6
    cx1 = L+R
    c1y = W - R - 25
    c2x = cx1
    c2y = c1y 
    r=R
    #imask0 = np.sum((INodes - np.vstack((C0x, C0y)).T )**2,axis=1) >= (r + mar)**2
    cimask1 = np.sum((INodes - np.array([cx1, c1y]))**2,axis=1) >= (r + mar)**2
    cimask2 = np.sum((INodes - np.array([a+L-R, c1y]))**2,axis=1) >= (r + mar)**2
    cimask3 = np.sum((INodes - np.array([a+L+L+R, c1y]))**2,axis=1) >= (r + mar)**2
    cimask4 = np.sum((INodes - np.array([a+L+L+a-R, c1y]))**2,axis=1) >= (r + mar)**2
    #imask0 =  np.sum((INodes - np.array([c2x, c2y]))**2,axis=1) >= (r + mar)**2
    #imask1 = INodes[INodes[:,:]>50]
    imask0 = (INodes[:,1] > b) | (INodes[:,0] < L) | (INodes[:,0] > L+a)
    
    imask1 = (INodes[:,1] > b) | (INodes[:,0] < L+L+a) | (INodes[:,0] > (W-L))
    
    #imask2 = (INodes[:,1] > b) | (INodes[:,0] < L+R) |  (INodes[:,0] > (W-L-R) ) | ( (INodes[:,0] > L+a) & (INodes[:,0] < L+a+L)  )
    imask2 = (INodes[:,1] > b) & (INodes[:,1] < (W-25)) &  (INodes[:,0] > L+R ) & (INodes[:,0] <a+ L-R ) 
    
    imask3 = (INodes[:,1] > b) & (INodes[:,1] < (W-25)) &  (INodes[:,0] < W - L-R ) & (INodes[:,0] >W-a- L+R ) 
    
    INodes = INodes[ imask0  & imask1 & imask1 & np.logical_not(imask2) & np.logical_not(imask3) & cimask2 & cimask1&cimask3 &cimask4]

    #INodes = INodes[cimask1]
    
    GNodes[:,0]=GNodes[:,0]-75
    INodes[:,0]=INodes[:,0]-75
    
    force = np.vstack((G1x-75, G1y)).T
    
    D0x = np.arange(0, L+g,  g)-75
    D0y = np.zeros_like(D0x) 
    
    D1x = np.arange(0, L+g,  g)+L+a-75
    D1y = np.zeros_like(D1x)     

    D2x = np.arange(0, L+g,  g)+L+a+L+a-75
    D2y = np.zeros_like(D2x) 
    
    fixed = np.vstack((np.vstack((D0x, D0y)).T,np.vstack((D1x, D1y)).T, np.vstack((D2x, D2y)).T )) 
 
    fixed_start = len(fixed_border)
    fixed_end = len(fixed_border) + len(down) 
    
    force_start =  len(G0x) +  len(fixed_border)
    force_end = len(G0x)+ len(G1x)  +  len(fixed_border) 
    return GNodes, INodes, force, fixed, fixed_start,fixed_end, force_start, force_end 
```

```python
def triarea(a, b, c):
    ab = b - a
    ac = c - a
    return 0.5 * (ab[:,0]*ac[:,1] - ab[:,1]*ac[:,0])
```

```python
def mass(Nodes, Elems, rho, t=0.01):
    Coords = Nodes[Elems]
    area = sum(list(map(triarea, Coords)))
    return rho * area * 10**(-4) * t
```

```python
def FEM(nodes, simplices, e, mu, t, p, fixed_elems, forced_elems):
    # расчет матрицы градиентов
    trng_vertices2 = nodes[simplices] # Пересчитаем, т.к. надо учитывать маски
    A = triarea(trng_vertices2[:,0], trng_vertices2[:,1], trng_vertices2[:,2]) # площади треугольников
    b = trng_vertices2[:, [1,2,0], 1] - trng_vertices2[:, [2, 0, 1], 1]
    c = trng_vertices2[:, [2,0,1], 0] - trng_vertices2[:, [1, 2, 0], 0]
    z = np.zeros_like(b[:,0])
    B1 = np.stack((b[:,0], z, b[:,1], z, b[:,2], z), axis=1)
    B2 = np.stack((z, c[:,0], z, c[:,1], z, c[:,2]), axis=1)
    B3 = np.stack((c[:,0], b[:,0], c[:,1], b[:,1], c[:,2], b[:,2]), axis=1)
    B = np.stack((B1, B2, B3), axis=1) # матрица градиентов
    #print(forced_elems)
    # расчет глобальной матрицы жесткости
    coef = 1/(2*A)
    B *= coef.reshape(-1, 1, 1)
    D = e / (1 - mu**2)*np.array([[1,mu,0], [mu,1,0], [0,0,0.5*(1-mu)]])
    BT = np.transpose(B, axes=(0,2,1))
    Ke = np.matmul(np.matmul(BT, D), B)*t*A.reshape(-1, 1, 1)
    coords0 = simplices[:, 0].reshape(-1, 1)
    coords1 = simplices[:, 1].reshape(-1, 1)
    coords2 = simplices[:, 2].reshape(-1, 1)
    EFT = np.hstack((coords0*2, coords0*2+1, coords1*2, coords1*2+1, coords2*2, coords2*2+1))
    Nn = nodes.shape[0]
    K = np.zeros((2*Nn, 2*Nn))
    for i in range(simplices.shape[0]):
        K[EFT[i, np.newaxis].T,EFT[i]] += Ke[i]
    dof = np.hstack((fixed_elems*2, fixed_elems*2+1))
    
    K[dof,:] = 0
    K[:,dof] = 0
    K[dof,dof] = 1
     
    # расчет силы, действующей на элементы
    F = np.zeros(2*Nn)
    edges_L = np.sqrt(np.sum((nodes[forced_elems[1:]] - nodes[forced_elems[:-1]])**2,axis=1))
    edges_F = np.ones((len(forced_elems),)) * p / edges_L[0]

    
    forced_elems_ext = np.append(np.append(forced_elems[-1], forced_elems), forced_elems[0]) 
    vecs = np.array([ 
        (nodes[i-1] - nodes[i+1]) / np.linalg.norm(nodes[i-1] - nodes[i+1]) for i in forced_elems_ext[1: -1] 
    ])
    
    
    alpha = -135*np.pi/180
    cos_alpha = np.cos(alpha)
    sin_alpha = np.sin(alpha)
    # cos
    F[forced_elems*2] += edges_F * vecs[:, 0]*cos_alpha
    
    # sin
    F[forced_elems*2+1] += edges_F * vecs[:, 1]*sin_alpha
    
    F[dof] = 0
    
    # расчет перемещений
    U = np.linalg.lstsq(K, F, rcond=None)[0]
    
 
    #print(U.shape, K.shape)
    # расчет напряжений и главных напряжений
    Ue = U[EFT, None] # матрица перемещений
    Eps = np.matmul(B, Ue) # тензор деформаций
    Sig = np.matmul(D, Eps) # тензор напряжений
    NodeA = np.zeros((Nn,1)) 
    NodeSig = np.zeros((Nn,3))
    for i, e in enumerate(simplices):
        NodeSig[e] += Sig[i].T*A[i]
        NodeA[e] += A[i]
    NodeSig /= NodeA
    NodeS1 = 0.5*(NodeSig[:,0]+NodeSig[:,1])+\
         np.sqrt((0.5*(NodeSig[:,0]-NodeSig[:,1]))**2+NodeSig[:,2]**2)
    
    return U, NodeS1
```

```python
def W(R, L, show=False): 
    GNodes, INodes, force, fixed, fixed_start,fixed_end, force_start, force_end  = detail_main_parts(R, L)
    
    Nodes = np.vstack((GNodes, INodes))
    
    trng = scipy.spatial.Delaunay(Nodes)
    
    Elems = trng.simplices.copy()
    
    mask = np.any(Elems >= Nodes.shape[0] - INodes.shape[0], axis=1)
    
    simplices = Elems[mask]

    fixed = np.array(range(fixed_start,fixed_end))
    forced = np.array(range( force_start, force_end))    

    #fixed = np.array(range(force_start, force_end))
    #forced = np.array(range(force_end, fixed_end))
    #print(forced)
    #print(fixed)
    W = 150
    
    U, NodeS1 = FEM(Nodes, simplices, e=2.2e7, mu=0.31, t=1, p=7000, fixed_elems=fixed, forced_elems=forced)
    scale = 500
    ResNodes = Nodes + scale*U.reshape(-1, 2)
    
    #if (show): plot_contour_map(ResNodes, NodeS1, GNodes, simplices, fixed, force)
    #border = GNodes
    #i_max = np.argmax(NodeS1)
    #i_min = np.argmin(NodeS1)
    #plt.figure(figsize=(12,12))
    #levels = np.linspace(NodeS1[i_min], NodeS1[i_max], 12)
    #handle = plt.tricontourf(ResNodes[:,0], ResNodes[:, 1], Elems[mask], NodeS1.ravel(), cmap='jet', levels=sorted(levels) )
    #plt.tricontour(ResNodes[:, 0], ResNodes[:, 1], simplices, NodeS1.ravel(), colors='k', levels=sorted(levels))
   


    #plt.plot(ResNodes[:, 0], ResNodes[:, 1], "o")
    
    
    
    
    #cb = plt.gcf().colorbar(handle)
    #cb.set_ticks(levels)
    #plt.tricontour(ResNodes[:, 0], ResNodes[:, 1], Elems[mask])
    #plt.plot(border[:,0],border[:,1],'.')    
 
    
    
    a=(W-3*L)/2.0
    b = 150-25-R
    
    S = 150*150 - 2*b*a -R*R*np.pi - R*(150-3*L-4*R)
 
    return max(NodeS1), S*0.01*70e-4
```

```python
mass_scale = np.abs(max(mass)-min(mass))
sig_scale = np.abs(max(sig)-min(sig))

mpar = [max(mass), min(mass)]
spar = [max(sig), min(sig)]


def scale(x,mx,mn):
    return (x-mn)/(mx-mn)


def opt_func(R1,R2,mp,sp,mpar,spar,lst=[]):
    #import pdb;pdb.set_trace()
    lst.append([R1, R2])
    if in_border(np.array([[R1, R2]]), points_bounds)[0]:
        s,m = WTF(R1, R2)
        m1 = scale(m, *mpar)
        mp1 = scale(mp, *mpar)
        s1 = scale(s, *spar)
        sp1 = scale(sp, *spar)
        return np.sqrt( (m1-mp1)**2 + (s1-sp1)**2 )
        #return np.abs(m-mp)/mpar[0]+np.abs(s-sp)/spar[0]
    else:
        return 1e3
    
    
perfect_points = np.array([[120.,80.]])

starting_points = np.array([[20, 90.]])

import pickle 
rez = []
for point, st_point in zip(perfect_points, starting_points):
    lst = []
    ret = scipy.optimize.minimize(lambda x, mp, sp, mpar,spar, lst: opt_func(x[0], x[1], mp, sp, mpar, spar, lst),
                                  st_point,
                                  args=(point[0],point[1],mpar,spar, lst),
                                  method='Nelder-Mead',
                                  options={'fatol':1e-10,'disp':True, 'maxiter':100})
    rez.append(ret['x'])
with open('rez.bin','wb') as mf:
    pickle.dump(rez,mf)
```
