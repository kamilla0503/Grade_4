Вариант 13: 
    
$x'' + \gamma x' + sin(x) = F sin(\omega t)  $


$ \gamma$ - коэффициент затухания, я взяла $ \gamma = 0.5$

F - амплитуда внешней силы. На данный момент я варьирую его, чтобы показать два фазовых портрета в разных режимах системы 

```python
omega_0 = 0
theta_0 = 0.2
g = l = 9.81
m = 1
n = 1000 #число шагов 
h = 0.05
omega_d = 2/3


q = 0.5
Fd = 0.5

```

Уравнение 2-го порядка, но легко приводится к системе первого порядка.

Численно решение получается с помощью метода Рунге-Кутты 4-го порядка (очень легко увидеть в функции rk_ste)

```python
#функции для дифференцирования 

def diff_omega(theta_i, omega_i, t):
    return -np.sin(theta_i) - q*omega_i + Fd*np.sin(t*omega_d)

def diff_theta(omega_i):
    return omega_i

def f(y, t):
    return np.array([diff_omega(y[1], y[0], t), diff_theta(y[0])])
```

```python
#метод для решения ДУ 

def runge_kutta(y, h, n, t): 
    for i in range(n-1):
        y[i+1] = rk_step(y, f, h, i, t)
    return y

def rk_step(y, f, h, i, t):
    k1 = h * f(y[i], t[i])
    k2 = h * f(y[i] + k1/2, t[i])
    k3 = h * f(y[i] + k2/2, t[i])
    k4 = h * f(y[i] + k3, t[i])
    
    return y[i] + k1/6 + k2/3 + k3/3 + k4/6

```

```python
def simulate():
    y = np.zeros((n,2))
    y[0, 1] = theta_0
    y[0, 0] = omega_0
    
    t = np.arange(0, n*h, h)
    y = runge_kutta(y, h, n, t)
    y[:,1] += np.pi 
    y[:,1] =  (y[:,1] + 2*np.pi)% (2*np.pi)
    y[:,1] -= np.pi
    return y, t
```

```python


def plot_this(y_1, y_2, t, labels, fig_name):
    plt.plot(t, y_1, color='b', linestyle='-', label=labels[0])
    plt.plot(t, y_2, color='b', linestyle=':', label=labels[1])
    
    plt.xlabel(labels[2])
    plt.ylabel(labels[3])
    plt.legend(bbox_to_anchor=(1.0, 1))
    plt.title(fig_name)

    plt.show()

def plot_dots(y_1, y_2, t, labels, fig_name):
    plt.plot(t, y_1, "*",  label=labels[0], markersize=2)
    plt.plot(t, y_2, linestyle="dashdot", color = 'k', label=labels[1], markersize=2)
    plt.xlabel(labels[2])
    plt.ylabel(labels[3])
    plt.legend(bbox_to_anchor=(1.0,1))
    plt.title(fig_name)
    plt.show()

```

Для получения двух фазовых портретов менялся параметр F - амплитуда внешней силы. 

Первый график показывает, что с течением времени фазовая траектория находится в ограниченном множестве и явно устойчива. 


Второй график показыает траекторию, которая не устоялась. Думаю, это пример аттрактора. 

```python
def graph_corr(x1):
    x1 = np.asarray(x1)
    x_avg = np.mean(x1)

    D = np.sum((x1 - x_avg)**2) / x1.shape[0]
    print('Дисперсия D = ', D)
    
    N = x1.shape[0]
    R_j = np.zeros(N)
    R_j_temp = np.zeros(N)
    r_j = np.zeros(N, np.float64)

    for j in range(int(N)):
        R_j_temp[j] = 0


        R_j[j] = np.sum((x1[:N-j] - x_avg) * (x1[j:] - x_avg))

        R_j[j] /= (N-j)

        r_j[j] = R_j[j] / D

    
    fig, ax = plt.subplots()

    #ax.xaxis.set_major_locator(ticker.MultipleLocator(500))
    fig.set_figwidth(12)
    fig.set_figheight(8)
    plt.plot(r_j[:5000])
    #plt.plot(r_j)
    
    #plt.plot(np.full(N, -1.0))
    #plt.plot(np.full(N, 1.0))

    plt.title('Корреляционная функция', fontsize=18) 
    plt.show()
```
