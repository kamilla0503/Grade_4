Мы рассматриваем классическую модель Изинга на квадратной решетке размера $(L \times L)$.Предполагаются, периодические граничные условия. Пусть $\sigma_{i,j}$ - это значения спинов на узлах решетки, принимающих значения $\pm 1$.  Энергия системы определяется как 
\begin{equation}
\frac{1}{2}\sum_{i=0}^{L-1}\sum_{j=0}^{L-1}E_{i,j}
\end{equation}
где
\begin{equation}
E_{i,j} = -\sigma_{i,j}(\sigma_{i+1,j} + \sigma_{i-1,j} + \sigma_{i,j+1} + \sigma_{i,j-1}).
\end{equation}
то есть каждый спин взаимодействует с 4 соседнями.  

Нам интересна намагниченность: 
\begin{equation}
m = \langle |\mu| \rangle
\end{equation}
\begin{equation}
\mu = \frac{1}{L^2} \sum_{i=0}^{L-1}\sum_{j=0}^{L-1}\sigma_{i,j}
\end{equation}

т.е. средняя сумма спинов. 

Для моделирование системы используется однокластерный метод Вольфа, реализованный на языке C++. 
https://github.com/kamilla0503/Grade_4/blob/master/Practicum/Ising/mcmc.cpp 


Для классической двумерной модели Изинга известно значения для критическое температуры: 
    $T_c = \frac{2}{\log(1+\sqrt{2} )}$ 

```python
T_critical = 2/np.log(1+np.sqrt(2))
J_critical = 1/T_critical
```

```python
data_np = np.loadtxt("Ising/longnear.txt", skiprows=0)
 
lengths = np.unique(data_np[:, 0])

fig = plt.figure(figsize=(13, 10))
ax = fig.add_subplot(111)

 
for l in lengths:
    if (l<250):
        continue
     
    data_l = data_np[data_np[:,0]==l]
    J_l = data_l[:, 1]
    
    bc = []
    errs=[]
    
    for j in range(len(J_l) ):
        e1 = norm.rvs(data_l[j, 7], data_l[j, 8], size = 1000)
        e2 = norm.rvs(data_l[j, 9], data_l[j, 10], size = 1000)
    
        cum = (e2 - e1*e1)*np.square(J_l[j])*l
        mean = np.mean(cum)
        
        err=np.std(cum)#/np.sqrt(len(cum))
        
        bc.append(mean)
        errs.append(err)

    plt.errorbar(J_l, bc, yerr=errs, fmt="s--", ms = 10, label ="L = " + str(np.sqrt(l)) )
 
 
plt.xlabel(r'$J$', fontsize = 22)
plt.ylabel(r"$ C  $ ", fontsize=22)
plt.tick_params(axis='x', labelsize=22) 
plt.tick_params(axis='y', labelsize=22) 

plt.xlim(0.32, 0.55)
plt.axvline(x=J_critical, label= "critical J")

plt.legend(loc="best", fontsize=20)
```
