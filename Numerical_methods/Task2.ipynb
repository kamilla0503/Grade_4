```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def task_function(x, y):
    return -x/(4*y)
```

```python
x = np.linspace(0, 2, 1500)
y_analytical_solution = np.sqrt(4-np.square(x))/2.0
```

```python
def euler(f, x, y, nsteps = 1000): 
    xs = np.linspace(0, 2,nsteps+1)
    ys = np.zeros(nsteps+1, dtype = float)
    xs[0] = x
    ys[0] = y

 
    h = 2/nsteps
    for i in range(1 , nsteps+1):

        ys[i] = ys[i-1] + h* f(xs[i-1], ys[i-1])
 
    return xs, ys, h
```

```python
x_sol1, y_sol1, h1 = euler(task_function, 0, 1)
x_sol2, y_sol2, h2 = euler(task_function, 0, 1, nsteps = 100)
x_sol3, y_sol3, h3 = euler(task_function, 0, 1, nsteps = 10)
x_sol4, y_sol4, h4 = euler(task_function, 0, 1, nsteps = 5)
```

```python
plt.figure(figsize=(12,7)) 
plt.xlabel('x', fontsize = 14)
plt.ylabel(r"$y$", fontsize=14) 
plt.plot(x, y_analytical_solution, '-', c = "k",label = "analytical solution")
plt.plot(x_sol1,y_sol1, '-', c = "magenta",label = "euler, h = %s"%(h1), alpha = 0.4, lw=3.5)
plt.plot(x_sol2,y_sol2, '-', c = "b",label = "euler, h = %s"%(h2), alpha = 0.4)
plt.plot(x_sol3,y_sol3, '-', c = "r",label = "euler, h = %s"%(h3), alpha = 0.5)
plt.plot(x_sol4,y_sol4, '-', c = "g",label = "euler, h = %s"%(h4), alpha = 0.5)

plt.legend(loc='best', fontsize = 14)
 
```

```python
def r_t_2(f, x, y, nsteps = 1000): 
    xs = np.linspace(0, 2,nsteps+1)
    ys = np.zeros(nsteps+1, dtype = float)
    xs[0] = x
    ys[0] = y

 
    h = 2/nsteps
    for i in range(1 , nsteps+1):
        k1 = h*f(xs[i-1], ys[i-1])
        k2= h*f(xs[i-1]+0.5*h, ys[i-1]+0.5*k1)
        
        ys[i] = ys[i-1] + k2
 
    return xs, ys, h
```

```python
x_sol_r2_1, y_sol_r2_1, h1 = r_t_2(task_function, 0, 1, 100)
x_sol_r2_2, y_sol_r2_2, h2 = r_t_2(task_function, 0, 1, nsteps = 10)
x_sol_r2_3, y_sol_r2_3, h3 = r_t_2(task_function, 0, 1, nsteps = 5)
```

```python
plt.figure(figsize=(12,7)) 
plt.xlabel('x', fontsize = 14)
plt.ylabel(r"$y$", fontsize=14) 
plt.plot(x, y_analytical_solution, '-', c = "k",label = "analytical solution", lw=3.5, alpha = 0.5)
plt.plot(x_sol_r2_1,y_sol_r2_1, '-', c = "magenta",label = "RK2, h = %s"%(h1), alpha = 0.4, lw=1.5)
plt.plot(x_sol_r2_2,y_sol_r2_2, '-', c = "aqua",label = "RK2, h = %s"%(h2), alpha = 0.6, lw=1.5)
plt.plot(x_sol_r2_3,y_sol_r2_3, '-', c = "orange",label = "RK2, h = %s"%(h3), alpha = 0.8, lw=1.5)
plt.legend(loc='best', fontsize = 14)
```

```python
def r_t_4(f, x, y, nsteps = 1000): 
    xs = np.linspace(0, 2,nsteps+1)
    ys = np.zeros(nsteps+1, dtype = float)
    xs[0] = x
    ys[0] = y

 
    h = 2/nsteps
    for i in range(1 , nsteps+1):
        k1 = h*f(xs[i-1], ys[i-1])
        k2= h*f(xs[i-1]+0.5*h, ys[i-1]+0.5*k1)
        k3= h*f(xs[i-1]+0.5*h, ys[i-1]+0.5*k2)
        k4= h*f(xs[i-1]+h, ys[i-1]+k3)
        ys[i] = ys[i-1] + (k1+2*k2+2*k3+k4 )/6.0
 
    return xs, ys, h
```

```python
x_sol_r4_1, y_sol_r4_1, h1 = r_t_4(task_function, 0, 1, 100)
x_sol_r4_2, y_sol_r4_2, h2 = r_t_4(task_function, 0, 1, nsteps = 10)
x_sol_r4_3, y_sol_r4_3, h3 = r_t_4(task_function, 0, 1, nsteps = 5)
```

```python
plt.figure(figsize=(12,7)) 
plt.xlabel('x', fontsize = 14)
plt.ylabel(r"$y$", fontsize=14) 
plt.plot(x, y_analytical_solution, '-', c = "k",label = "analytical solution", lw=3.5, alpha = 0.5)
plt.plot(x_sol_r4_1,y_sol_r4_1, '-', c = "magenta",label = "RK4, h = %s"%(h1), alpha = 0.4, lw=1.5)
plt.plot(x_sol_r4_2,y_sol_r4_2, '-', c = "aqua",label = "RK4, h = %s"%(h2), alpha = 0.6, lw=1.5)
plt.plot(x_sol_r4_3,y_sol_r4_3, '-', c = "orange",label = "RK4, h = %s"%(h3), alpha = 0.8, lw=1.5)
plt.legend(loc='best', fontsize = 14)
```
