###  1. Density function: 


$ p(x) =\frac{1}{{\sigma \sqrt {2\pi } }} e^{  - ( {x - \mu } )^2 / ( 2\sigma ^2 )   }   $


$\mu = -13.5 $

$\sigma = 4 $

 $ p(x) =\frac{1}{{ 2 \sqrt {2\pi } }} e^{  - ( {x + 13.5 } )^2 / 8   }   $


```python
def p(x, mu = -13.5, sigma = 2):
    p = np.exp(-np.square(x-mu)/2/np.square(sigma))/sigma/np.sqrt(2*np.pi)
    return p
```

### 4 Probabilities

```python
from scipy.stats import norm, chi2, f, t, binom_test
from scipy.special import kolmogorov, erf
rv = norm(-13.5, 2)

p1 = rv.cdf(-20) - rv.cdf(-21)
p1
```

$P(-21<Z<-20) \approx 0.0005 $

```python
p2 = rv.cdf(-13) - rv.cdf(-17)
p2
```

$P(-17<Z< - 13) \approx 0.558647 $

$P(-17<Z<-13) >  P(-21<Z<-20) $

```python
x = np.linspace(-21.5, -5.5, 200)

p_x = p(x)

 
plt.figure(figsize=(10,5.5)) 
plt.xlabel("x", fontsize=15)
plt.ylabel("p(x)", fontsize=15)
plt.plot(x, p_x, label = r"$Z$") 
 
plt.fill_between(np.linspace(-17,-13,20), rv.pdf(np.linspace(-17,-13,20)), alpha =0.5, color="purple", label = "[-17,-13]" )


plt.fill_between(np.linspace(-21,-20,20), rv.pdf(np.linspace(-21,-20,20)), alpha =0.5, color="b", label = "[-21,-20]" )


plt.legend(loc='best',fontsize=14)
 
```
