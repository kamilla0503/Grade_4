# Проверка решения.

Решение было получено на 2 процессах, 10000 итераций на 100 точках 

```python
data = np.loadtxt("HEAT_EQUATION_2_100.txt", skiprows=1)t
```

```python
plt.plot(data[:,0::2], data[:,1::2], "s",label = "Numerical solution")
```

```python
t = 5.10152e-05*10000 


X = data[:,0]
 
 
#это решение, которое должно получаться методом Фурье (но я его подбирала)
Z = np.exp(-np.pi*np.pi*t)*np.sin(np.pi*X)/np.pi*4 

plt.plot(data[:,0::2], data[:,1::2], "s",label = "Numerical solution")

plt.plot(X, Z, c = "k", label = "Analytical solution")

plt.legend()
```
