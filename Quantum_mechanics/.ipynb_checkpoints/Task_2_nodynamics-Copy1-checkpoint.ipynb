# Прохождение через прямоугольный барьер

Тестовый пример + сравнение с известной формулой

```python
x = np.linspace(0,1,200)
L = x[-1]-x[0]
dx= x[1]-x[0]
Nx = len(x)  
V = np.zeros(Nx)   

V[int(0.45*Nx):int(0.55*Nx)] = 100

E = np.arange(101, 400)*L*L
r = np.zeros(len(E))
t = np.zeros(len(E))

t_a =  np.zeros(len(E), dtype=complex)
T_a =  np.zeros(len(E))

#fig, ax = plt.subplots(figsize=(10, 6))
 
for i in range(len(E)):
    V = np.zeros(Nx)   
    V[1:-1] = 100 

    #V[int(0.45*Nx):int(0.55*Nx)] = 100/L/L

    H = scipy.sparse.diags([1, -2, 1], [-1, 0, 1], shape=(Nx, Nx)).toarray() + ( E[i] - np.diag(V)  )*dx*dx
    u = H.diagonal()
    u = -2  + dx*dx*(E[i]-V )
    R = np.zeros(len(x), dtype=complex)
    
    R[-1] = -1/(u[-1]/2 + 1j*dx*np.sqrt(E[i]) )
    
    for j in range (len(x)-2, -1, -1):
        R[j] = -1/(u[j+1]+R[j+1])
        
    phi = np.zeros(len(x)+1, dtype=complex)
    
    phi[0] = 2* 1j*dx*np.sqrt(E[i])/(R[0]+ u[0]/2 + 1j*dx*np.sqrt(E[i]))
    #print(phi[0], R[0], u[0])
    
    for j in range(1, len(phi)):
        phi[j]=R[j-1]*phi[j-1]
        
    t[i] = np.square(np.abs(phi[-1]))
    r[i] = np.square(np.abs(-1+phi[0]))
    
    ki = np.sqrt(E[i])/ ( np.sqrt(E[i] - V[len(V)//2]))
    theta=L*( np.sqrt(E[i] - V[len(V)//2]) )
    
    t_a[i] = 4*ki/ (np.square(1+ki) * np.exp(-1j*theta) - np.square(1-ki) * np.exp(1j*theta) )
    
    #if (i%150==0):
        #plt.plot(x, np.square(np.abs(phi[1:])), label = "E = "+str(E[i]))
    
    
T_a = np.square(np.abs(t_a))
#plt.plot(x, V)    
#plt.legend()
    #s = np.linalg.solve(H,np.zeros(len(x)))
    #print(H)
    #print(s)
    #break 
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

plt.plot(E,t, "*", label = r'$T$, numerical result' )

plt.plot(E,T_a,  c = "k", alpha = 0.5,label = r'Analytical, 4.25 (p.44)' )
#plt.plot(E, r,  label = r'$R$' )
plt.xlabel(r'$E$', fontsize = 14)
plt.ylabel(r'$T(E)$', fontsize = 14)
 
plt.legend()
```
