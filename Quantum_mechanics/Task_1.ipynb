```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd

import scipy.linalg as linalg
from scipy.linalg import ldl
import scipy.sparse.linalg
import scipy.sparse

import matplotlib.animation as animation
import matplotlib.cm as cm

%matplotlib notebook
```

# Свободное движение 

$ V(x) = 0$

$N=20$

```python
x = np.linspace(-1,1,20)
dx=2/20
Nx = len(x)  
V = np.zeros(Nx)  #V=0, сравниваем свободное движение 
```

```python
H = scipy.sparse.diags([-1, 2, -1], [-1, 0, 1], shape=(Nx, Nx)).toarray() + np.diag(V) 

print("Матрица Гамильтониана")
for i in range(len(H)):
    print("%.1f "*len(H[i]) % tuple(H[i]) )
    
    
```

```python
eigs = linalg.eigvals(H ) #для показа собственных значений 
eivs = linalg.eigh(H)[1]
print("Список собственных значений")
print(eigs)

fig, ax = plt.subplots(figsize=(10, 6))
plt.plot(sorted(eigs), "*" )


L = 2
#plt.plot(2*(1-np.cos(np.pi/2*np.arange(Nx)*dx)))



plt.xlabel(r'N ', fontsize = 22)
plt.ylabel(r"E_n", fontsize=22)
plt.title("График собственных значений")
plt.tick_params(axis='x', labelsize=22) 
plt.tick_params(axis='y', labelsize=22) 
```

## Потенциальная яма 


$N = 1000 $

```python
x = np.linspace(-1,1,1000)
dx=2/1000
Nx = len(x)  

V = np.zeros(Nx)

V[int(0.45*Nx):int(0.55*Nx)] = -2

H = scipy.sparse.diags([-1, 2, -1], [-1, 0, 1], shape=(Nx, Nx)).toarray() + np.diag(V) 


fig, ax = plt.subplots(figsize=(10, 6))
plt.plot(x, V)
plt.ylabel(r"V", fontsize=22)
```

```python
eigs = linalg.eigvals(H ) #для показа собственных значений 
eivs = linalg.eigh(H)[1]

fig, ax = plt.subplots(figsize=(10, 6))
plt.plot(sorted(eigs), "*" )
 
plt.plot([0, Nx], [0, 0], c="k")

plt.xlabel(r'N ', fontsize = 22)
plt.ylabel(r"E_n", fontsize=22)
plt.title("График собственных значений")
plt.tick_params(axis='x', labelsize=22) 
plt.tick_params(axis='y', labelsize=22) 
```
