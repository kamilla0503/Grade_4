```python
import pandas as pd
import scipy
from statsmodels.stats.weightstats import *
import statsmodels.stats.multitest as smm
```

```python
data = pd.read_csv('gene_high_throughput_sequencing.csv', sep =',')
```

```python
data_cancer = data[data["Diagnosis"]=="cancer"]
data_cancer.shape
```

```python
data_normal =  data[data["Diagnosis"]=="normal"]
data_early =  data[data["Diagnosis"]=="early neoplasia"]
data_early.shape
```

```python
gene_names = list(filter(lambda x: x!="Patient_id" and x!="Diagnosis" ,list(data)) )

 
```

```python
corr_norm_early = []
corr_canc_early = []

i=0
for t in gene_names:
    p =  scipy.stats.ttest_ind(data_normal[t], data_early[t], equal_var = False)[1]
    corr_norm_early.append([t, p])    
    p =  scipy.stats.ttest_ind(data_early[t], data_cancer[t], equal_var = False)[1]
    
    corr_canc_early.append([t, p])
    print(i, end = " ")
    i=i+1
```

```python
correlation_n_e = pd.DataFrame.from_records(corr_norm_early)
correlation_c_e = pd.DataFrame.from_records(corr_canc_early )
```

```python
def g_c(T, C):
    if(T>C):
        return T/C
    else:
        return -C/T
```

```python
correlation_n_e["holm"] = new_n_e
```

```python
correlation_n_e["really_matters"] = correlation_n_e["holm"]<0.025
```

```python
genes_n_e =  list(correlation_n_e[correlation_n_e["really_matters"]][0])
```

```python
f_n_e = []
for name in genes_n_e:
    f_n_e.append(g_c(data_early[name].mean(),data_normal[name].mean()))
```
